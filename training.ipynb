Parameters:

rotation_range (40): This controls the range of random rotations applied to the image, in degrees. For plant leaves, a lower value (10-20 degrees) might be more suitable, as excessive rotation can distort the natural orientation of leaves.
width_shift_range (0.2): This controls the amount of random horizontal shift applied to the image, as a fraction of the image width. A lower value (0.1 or less) might be better for leaves, as large shifts can detach them from the stem or background.
height_shift_range (0.2): Similar to width shift, controls the vertical shift. A lower value (0.1 or less) is recommended for similar reasons.
shear_range (0.2): This parameter controls the amount of random shearing applied, which can distort the shape of the image. For leaves, consider removing it entirely (0.0) as it might cause unnatural deformations.
zoom_range (0.2): This controls the range of random zoom applied. While some zoom can be beneficial, a lower value (0.1 or less) is advisable to avoid excessive magnification or shrinking, which can affect leaf details.
horizontal_flip (True): This flips the image horizontally. This can be beneficial for leaves as it introduces variability without significantly altering their structure.
fill_mode ('nearest'): This defines how to fill the background pixels during transformations. While 'nearest' might work, consider using 'constant' and setting a color close to the background foliage for a more natural look.

```python
from keras.preprocessing.image import ImageDataGenerator
import os
import numpy as np
from PIL import Image


def augment_and_save(image, save_folder, num_images=20):
    # Create ImageDataGenerator object with desired augmentation parameters
    datagen = ImageDataGenerator(
        rotation_range=5,
        width_shift_range=1,
        height_shift_range=1,
        zoom_range=0.1,
        horizontal_flip=True,
        vertical_flip=True,
    )

    # Ensure save folder exists, if not create it
    if not os.path.exists(save_folder):
        os.makedirs(save_folder)

    # Reshape the image to match the expected input shape of the ImageDataGenerator
    image = np.array(image)  # Convert PIL image to numpy array
    image = image.reshape((1,) + image.shape)  # Add batch size dimension

    # Generate and save augmented images
    for i, batch in enumerate(datagen.flow(image, batch_size=1, save_to_dir=save_folder,
                                            save_prefix='augmented', save_format='jpg')):
        if i >= num_images:
            break


# # Example usage

# main_directory_path = './data/train_r/Apple leaf_3.jpg'
# # Read image using PIL
# img = Image.open(main_directory_path)
# augment_and_save(img, './data/augument/Apple leaf_3', num_images=10)
# make it read from the train_r all images and create
directory_names = os.listdir('./data/train_r')
for i in directory_names:
    directory_path = './data/train_r/' + i
    img=Image.open(directory_path)
    augment_and_save(img, './data/augument/'+i, num_images=10)
```
