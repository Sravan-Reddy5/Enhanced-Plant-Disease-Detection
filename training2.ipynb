```python
# resize the images
import cv2
import os
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt

# Specify the directory path
main_directory_path = './data/train_for_final'
resize_directory_path = './data/train_for_final_resized'

directory_names = os.listdir(main_directory_path)
print(directory_names)

for i in directory_names:
    directory_path = main_directory_path + '/' + i
    file_names = os.listdir(directory_path)
    for j in file_names:
        file_path = directory_path + '/' + j
        img = cv2.imread(file_path)
        img = cv2.resize(img, (224, 224))
        save_path = resize_directory_path + '/' + i
        if not os.path.exists(save_path):
            os.makedirs(save_path)
        save_path = save_path + '/' + j
        cv2.imwrite(save_path, img)
        
```

```python
#create the X array
import os
import numpy as np
from PIL import Image
import cv2
main_directory_path = './data/train_for_final_resized'
directory_names = os.listdir(main_directory_path)
print(directory_names)
X=[]
for i in directory_names:
    directory_path = main_directory_path + '/' + i
    file_names = os.listdir(directory_path)
    for j in file_names:
        file_path = directory_path + '/' + j
        img = cv2.imread(file_path)
        X.append(img)
```

```python
X
```

```python
X=np.array(X)
```

```python
X.shape
```

```python
import tensorflow as tf
import tensorflow_hub as hub
from tensorflow import keras
from tensorflow.keras.models import Sequential
```

```python
num_classes = 13
model1 = tf.keras.Sequential([
    hub.KerasLayer("https://www.kaggle.com/models/google/mobilenet-v2/frameworks/TensorFlow2/variations/tf2-preview-feature-vector/versions/4", output_shape=[1280],
                   trainable=True),  # Can be True, see below.
    tf.keras.layers.Dense(num_classes, activation='softmax')
])
model1.build([None, 224, 224, 3])  # Batch input shape.
```

```python
model1.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model1.summary()
```

```python
history = model1.fit(X, Y, epochs=10, validation_split=0.2)
```

```python
# accuracy of the model
from sklearn.metrics import accuracy_score
y_pred = model1.predict(X)
y_pred = np.argmax(y_pred, axis=1)
accuracy = accuracy_score(Y, y_pred)
print('Accuracy: %f' % accuracy)
```
